### 1. Para onde foram os refugiados ucranianos?

```python
refugees_file_path = "../Dados/UKR_refugee_by_country.csv"

dados_refugiados = pd.read_csv(refugees_file_path)

display(dados_refugiados.describe())

display(dados_refugiados)
```

```python
dados_ref_max = dados_refugiados.groupby('country').individuals.max()
dados_ref_min = dados_refugiados.groupby('country').individuals.min()
dados_ref_mean = dados_refugiados.groupby('country').individuals.mean()

print(dados_ref_max, dados_ref_min, dados_ref_mean, sep='\n\n')
```
