```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
refugees_filepath = "../Dados/UKR_refugee_by_countries.csv"

dados_refugiados = pd.read_csv(refugees_filepath)

dados_refugiados.describe()
```

```python
dados_refugiados.head(10)
```

```python
# selecionar primeiras 6 linhas
dados_ref_final = dados_refugiados.iloc[0:6]

dados_ref_final.head(6)
```

```python
# selecionar últimas 6 linhas
dados_ref_ini = dados_refugiados.iloc[-6:]

dados_ref_ini.head(6)
```

```python
# dados agrupados por país
dados_ref_max = dados_refugiados.groupby('country').individuals.max()
dados_ref_min = dados_refugiados.groupby('country').individuals.min()
dados_ref_mean = dados_refugiados.groupby('country').individuals.mean()

print(dados_ref_max, dados_ref_min, dados_ref_mean, sep='\n\n')
```

```python
dados_ref_stats = dados_refugiados.groupby('country').individuals.agg([len, min, max])

dados_ref_stats.head(10)
```

```python
from typing import List, TypedDict

class CountryRef(TypedDict):
    qtd: int
    date: str

class Country(TypedDict):
    name: str
    lista: List[CountryRef]
    
dados_country_ref_by_date = dados_refugiados[['country', 'date', 'individuals']]

paises = dict()

country_aux = dados_country_ref_by_date.country.tolist()
date_aux = dados_country_ref_by_date.date.tolist()
individual_aux = dados_country_ref_by_date.individuals.tolist()

for country, date, individuals in zip(country_aux, date_aux, individual_aux):    
    if country in paises.keys():
        
        paises[country]['lista'].insert(0, 
            CountryRef(
                qtd = individuals,
                date = date
            )
        )
    else:
        paises[country] = Country(
            lista = [
            CountryRef(
                qtd = individuals,
                date = date
            )]
        )
    
plt.ticklabel_format(style='plain')

for pais in paises.keys():
    individuals = [indiv['qtd'] for indiv in paises[pais]['lista']]
    dates = [el['date'] for el in paises[pais]['lista']]
    
    plt.plot(dates, individuals, 'b')
    
    plt.title(f'Número de refugiados em {pais}')
    
    plt.show()


```

```python
countries = dados_ref_stats.index.tolist()
max_by_country = dados_ref_stats['max'].tolist()

plt.ticklabel_format(style='plain')

plt.barh(countries, max_by_country)

plt.show()
```
