```python
import numpy as np
import pandas as pd
import re
from matplotlib import pyplot as plt
from nltk.corpus import stopwords
```

```python
twitter_data_file_path = "../Dados/Ukr_war_tweets.csv"

df_twitter = pd.read_csv(twitter_data_file_path, dtype = 'unicode')

display(df_twitter)
```

```python
attrs = ['Datetime', 'Text', 'Language', 'hashtag']

df_twitter_texts = df_twitter[attrs]

df_twitter_texts.dropna(subset = ['Text'])

display(df_twitter_texts.head())
```

```python
def get_words_list(string: str) -> list:
    pattern = r"\b[A-Za-z#]+"
        
    return re.findall(pattern, string.lower()) 
    
def count_word_frequency(df: pd.DataFrame) -> dict:
    word_freq = dict()
    
    for idx, row in df.iterrows():
        tweet_words = get_words_list(str(row['Text']))
        
        for word in tweet_words:
            if word in word_freq.keys():
                word_freq[word] += 1
            else:
                word_freq[word] = 1
                
    return word_freq

def get_word_freq(word_frequency: dict, min_val: int = 1) -> dict:
    stop = set(stopwords.words('english'))
    languages = set(('russian', 'french', 'spanish', 'portuguese'))
    
    for language in languages:
        stop = stop.union(set(stopwords.words(language)))    
        
    words_list = set(filter(lambda x: ((word_frequency[x] >= min_val) & (x not in stop)), word_frequency.keys()))
    
    return {key: word_freq[key] for key in words_list}
```

```python
a = "Oi eu fui no parque weee AAAA weee eu oi oi Oi oi Oi if eu aqui else eu ali"

print(x := get_words_list(a))

word_freq = dict()
        
for word in x:
    if word in word_freq.keys():
        word_freq[word] += 1
    else:
        word_freq[word] = 1
        
wf = get_word_freq(word_freq)
print(word_freq)
print(wf)
```

```python
teste = count_word_frequency(df_twitter_texts.head(15))

print(teste, sep = '\n')
```

```python
word_freq = count_word_frequency(df_twitter_texts)

print(len(word_freq))
```

```python
word_freq_least100 = get_word_freq(word_freq, 1000)

print((word_freq_least100))
```
